#### PIL Library의 Image Library는 Image Load/Save 등에 사용되는 Library이다
#### pylab은 data/image의 plot 및 분석 등에 사용되는 library이다

```python
from PIL import Image
import numpy as np
import pylab as plt
```

#### Image Library의 open을 이용해서 Image를 불러올 수 있고
#### convert('L')을 사용하면 흑백사진으로 불러온다 (8비트)
#### 이것을 np.array로 감싸 주면 행렬 형태로 변환할 수 있다

```python
img = np.array(Image.open('D:\\ipy\\cameraman2.tif').convert('L'))
```

#### MATLAB교재에 있는 imadd를 구현해야 한다. imadd의 특징은 덧셈 후 결과를 0-255로 만들어 주는 것이다
#### 하지만 Image Library를 사용한 뒤 numpy array로 바꾸면 uint8 형식이 되기 때문에
#### 연산 과정에서 255가 넘어가면 자동으로 overflow가 발생하기 때문에
#### 그것을 방지하기 위해 연산 중간에는 float나 double(실수형)으로 바꾸어 준다

```python
img = img.astype(dtype='float')
```

Image Add 연산 후 0-255로 clip

```python
img2 = np.clip(img+75,0,255)
```

#### 창을 띄운다

```python
plt.figure()
```

아래 두 함수는 MATLAB의 그것과 동일

```python
plt.subplot(1,2,1)
plt.imshow(img)
```

```python
plt.gray() # 흑백 Plot
plt.axis('off') # 축을 제거해준다
plt.title('Original Image') # Image Title

plt.subplot(1,2,2)
plt.imshow(img2)
plt.gray()
plt.axis('off')
plt.title('Brighter Image')

plt.show() # Python에서는 실제로 Show()를 해야 창이 띄워진다
```
