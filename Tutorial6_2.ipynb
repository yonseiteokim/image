```python
%matplotlib notebook
from PIL import Image
import numpy as np
import pylab as plt
from roipoly import roipoly #MATLAB의 roipoly과 같은 역할을 해주는 외부 Python Code이다
```

pout.tif 이미지를 불러온다 (I)

```python
I = np.array(Image.open('D:\\ipy\\pout.tif').convert('L'))

plt.imshow(I)
plt.gray()
plt.axis('off')
# 좌 클릭으로 도형을 그리고, 우 클릭으로 그리기를 끝낸다 (Notebook에서 사용하면 매우 느릴 수 있음)
ROI1 = roipoly(roicolor='r')
```

```python
plt.imshow(I)
plt.axis('off')
# 그린 도형을 겹쳐서 Plot해 본다
ROI1.displayROI()
plt.show()
```

```python
# ROI(Region of Interest)의 Mask를 가져온다
bw2 = ROI1.getMask(I)
# 이 Mask은 True/False의 Boolean 형태이므로 0 아니면 255값이 되도록 바꿔준다
bw2 = 255*bw2.astype(dtype='uint8')
plt.figure(2)
plt.imshow(bw2)
plt.axis('off')
plt.show()
```

```python
I2 = np.bitwise_and(I,bw2)
plt.figure(3)
plt.subplot(1,2,1)
plt.imshow(I)
plt.axis('off')
plt.title('Original Image')
plt.subplot(1,2,2)
plt.imshow(I2)
plt.axis('off')
plt.title('Bit AND Result')
plt.show()
```

```python
# numpy에서 complement는 invert를 사용한다
bw_cmp = np.invert(bw2)
plt.figure(4)
plt.imshow(bw_cmp)
plt.axis('off')
plt.show()
```

```python
# bitwise_or도 마찬가지이다
I3 = np.bitwise_or(I,bw2)
plt.figure(5)
plt.subplot(1,2,1)
plt.imshow(I)
plt.axis('off')
plt.title('Original Image')
plt.subplot(1,2,2)
plt.imshow(I3)
plt.axis('off')
plt.title('Bit OR Result')
plt.show()
```
