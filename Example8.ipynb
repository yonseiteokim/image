# Image Study Example 8

```python
import numpy as np

input1 = np.array([[20,40,0],[178,198,64],[77,128,1]])
print(input1)
```

#### Zero Vector 256 Size로 LUT(Look up table)을 만든다

```python
LUT = np.zeros((256), dtype='uint8')
print(LUT)
```

#### Example 8-5에 나와있는대로 LUT를 만든다
#### Python은 index가 0부터 시작 (MATLAB은 1부터 시작)
#### 또한 [0:65]면 0부터 64까지의 index를 의미한다는 것에 주의한다
#### arange함수는 (0,65,1)일 경우 0부터 64까지 1 간격으로 벡터를 생성하는 것이다
#### 즉 [0, 1, 2, 3, ... 64]를 의미한다

```python
LUT[0:65] = 2*np.arange(0,65,1)
LUT[65:129] = 128
LUT[129:256] = np.arange(130,257,1)-1
print(LUT)
```

#### Look Up Table을 통과시키는 방법은 아래와 같이 간단히 할 수 있다

```python
print(LUT[input1])
```

#### Image도 통과시켜 보자

```python
from PIL import Image
I = np.array(Image.open('D:\\ipy\\klcc_a.png').convert('L'))
O = LUT[I]
```

```python
import pylab as plt

plt.figure(1)
plt.subplot(1,2,1)
plt.gray()
plt.imshow(I)
plt.axis('off')
plt.title('Input Image')

plt.subplot(1,2,2)
plt.gray()
plt.imshow(O)
plt.axis('off')
plt.title('Output Image')

plt.show()
```
