#### 새롭게 추가된 함수는 plt.hist()라는 함수인데, 히스토그램을 바로 Plot시키는 함수이다
#### 여기에 처음으로 img.flatten()이 들어가는데 이 함수는 2-D Array를 1-D 로 만들어준다
#### 뒤에 들어가는 숫자 256은 Histogram의 Bin의 개수이다

```python
from PIL import Image
import pylab as plt
import numpy as np

img = np.array(Image.open('D:\\ipy\\pout.tif').convert('L'))

plt.figure(1)
plt.subplot(2,1,1)
plt.gray()
plt.axis('off')
plt.imshow(img)

plt.subplot(2,1,2)
plt.hist(img.flatten(), 256)

plt.show()
```

### Bin을 64개쓰면 아래처럼 된다

```python
plt.figure(1)
plt.hist(img.flatten(), 64)

plt.show()
```

#### 다른 Image의 Histogram이다
#### Image의 Contrast 가 낮으면 Histogram이 몰려 있고
#### Contrast(대비)가 높으면 Histogram이 퍼져 있는 것을 확인할 수 있다

```python
img = np.array(Image.open('D:\\ipy\\salzburg.png').convert('L'))

plt.figure(1)
plt.subplot(2,1,1)
plt.gray()
plt.axis('off')
plt.imshow(img)

plt.subplot(2,1,2)
plt.hist(img.flatten(), 256)

plt.show()
```
